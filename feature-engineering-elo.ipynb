```python
import numpy as np 
import pandas as pd 
import os
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
print(os.listdir("../input"))
import gc
```

```python
train = pd.read_csv('../input/train.csv')
test = pd.read_csv('../input/test.csv')
```

```python
def add_historical_features(df):
    his_tran = pd.read_csv('../input/historical_transactions.csv')
    his_tran['authorized_flag'] = his_tran['authorized_flag'].apply(lambda x: 1 if x == 'Y' else 0)
    t_df = pd.merge(df,pd.DataFrame(his_tran.groupby(['card_id']).count()['purchase_amount']),on='card_id',how='left')
    t_df = t_df.rename({'purchase_amount':'purchase_count'}, axis='columns')
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id']).sum()['installments']),on='card_id',how='left',suffixes=('a','b'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id']).mean()['installments']),on='card_id',how='left',suffixes=('c','d'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id']).std()['installments']),on='card_id',how='left',suffixes=('e','f'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id','authorized_flag']).count()['purchase_amount'].unstack()),on='card_id',how='left',suffixes=('g','h'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id','authorized_flag']).mean()['purchase_amount'].unstack()),on='card_id',how='left',suffixes=('i','j'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id','authorized_flag']).std()['purchase_amount'].unstack()),on='card_id',how='left',suffixes=('k','l'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id']).count()['city_id']),on='card_id',how='left')
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id']).sum()['purchase_amount']),on='card_id',how='left',suffixes=('m','n'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id']).mean()['purchase_amount']),on='card_id',how='left',suffixes=('o','p'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id']).std()['purchase_amount']),on='card_id',how='left',suffixes=('q','r'))
    his_tran['category_1'] = his_tran['category_1'].astype('category')
    his_tran['category_1'] = his_tran['category_1'].cat.codes
    t_df = pd.merge(t_df,pd.DataFrame(his_tran[['card_id','category_1']].groupby(['card_id']).max()),on='card_id',how='left',suffixes=('s','t'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran[['card_id','category_1']].groupby(['card_id']).min()),on='card_id',how='left',suffixes=('u','v'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id','category_1']).count()['authorized_flag'].unstack()),on='card_id',how='left',suffixes=('w','x'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id','category_1']).mean()['authorized_flag'].unstack()),on='card_id',how='left',suffixes=('y','z'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id','category_1']).std()['authorized_flag'].unstack()),on='card_id',how='left',suffixes=('_a','_b'))
    his_tran['category_2'] = his_tran['category_2'].astype('category')
    his_tran['category_2'] = his_tran['category_2'].cat.codes
    t_df = pd.merge(t_df,pd.DataFrame(his_tran[['card_id','category_2']].groupby(['card_id']).max()),on='card_id',how='left',suffixes=('_c','_d'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran[['card_id','category_2']].groupby(['card_id']).min()),on='card_id',how='left',suffixes=('_e','_f'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id','category_2']).count()['authorized_flag'].unstack()),on='card_id',how='left',suffixes=('_g','_h'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id','category_2']).mean()['authorized_flag'].unstack()),on='card_id',how='left',suffixes=('_i','_j'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id','category_2']).std()['authorized_flag'].unstack()),on='card_id',how='left',suffixes=('_k','_l'))
    his_tran['category_3'] = his_tran['category_3'].astype('category')
    his_tran['category_3'] = his_tran['category_3'].cat.codes
    t_df = pd.merge(t_df,pd.DataFrame(his_tran[['card_id','category_3']].groupby(['card_id']).max()),on='card_id',how='left',suffixes=('_m','_n'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran[['card_id','category_3']].groupby(['card_id']).min()),on='card_id',how='left',suffixes=('_o','_p'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id','category_3']).count()['authorized_flag'].unstack()),on='card_id',how='left',suffixes=('_q','_r'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id','category_3']).mean()['authorized_flag'].unstack()),on='card_id',how='left',suffixes=('_s','_t'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id','category_3']).std()['authorized_flag'].unstack()),on='card_id',how='left',suffixes=('_u','_v'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id']).sum()['month_lag']),on='card_id',how='left',suffixes=('_w','_x'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id']).mean()['month_lag']),on='card_id',how='left',suffixes=('_y','_z'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id']).std()['month_lag']),on='card_id',how='left',suffixes=('_a1','_b1'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id','month_lag']).count()['authorized_flag'].unstack()),on='card_id',how='left',suffixes=('_a2','_b2'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id','month_lag']).mean()['authorized_flag'].unstack()),on='card_id',how='left',suffixes=('_a3','_b3'))
    t_df = pd.merge(t_df,pd.DataFrame(his_tran.groupby(['card_id','month_lag']).std()['authorized_flag'].unstack()),on='card_id',how='left',suffixes=('_a4','_b4'))
    return t_df
```

```python
train_df = add_historical_features(train)
train_df.head()
```

```python
train_df.fillna(0,inplace=True)
```

```python
test_df = add_historical_features(test)
test_df.head()
```

```python
test_df.fillna(0,inplace=True)
```

```python
features = train_df.columns.drop(['first_active_month','card_id','target']).values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train_df[features], train_df['target'], test_size = 0.25, random_state = 0)
```

```python
import lightgbm as lgb
from sklearn.metrics import mean_squared_error
```

```python
d_train = lgb.Dataset(x_train, label=y_train)
```

```python
params = {}
params['learning_rate'] = 0.005
params['boosting_type'] = 'gbdt'
params['objective'] = 'regression'
#params['bagging_fraction'] = 0.7
#params['bagging_freq'] = 100
params['metric'] = 'rmse'
params['sub_feature'] = 0.78
params['num_leaves'] = 87
params['min_data'] = 488
params['max_depth'] = 33
```

```python
clf = lgb.train(params, d_train, 2000)
```

```python
y_pred = clf.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
print(mse**0.5)
```

```python
y_pred2 = clf.predict(x_train)
mse2 = mean_squared_error(y_train,y_pred2)
print(mse2**0.5)
```

```python
test_df['target'] = clf.predict(test_df[features])
```
